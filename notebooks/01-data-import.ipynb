```python
# scripts/import_data.py
import requests
import pandas as pd
from io import BytesIO
import os

# Define paths based on project structure
RAW_DATA_DIR = "data/raw/"
os.makedirs(RAW_DATA_DIR, exist_ok=True)  # Create directory if it doesn't exist

# KoBoToolbox credentials and export URL
ASSET_UID = "aTVFPZ5vqTEWBJzNbzUftT"
USERNAME = "alkasemh"
PASSWORD = "srd#2023"
EXPORT_URL = f"https://eu.kobotoolbox.org/api/v2/assets/{ASSET_UID}/export-settings/esViNEBj8toYRPBhomYxcfs/data.xlsx"

# File path for saving raw data
RAW_FILE_PATH = os.path.join(RAW_DATA_DIR, "srd_patient_exit_interviews_raw.xlsx")

# Get the Excel file
response = requests.get(EXPORT_URL, auth=(USERNAME, PASSWORD))

if response.status_code == 200:
    print("‚úÖ Export downloaded successfully!")
    
    # Save raw Excel file to data/raw/
    with open(RAW_FILE_PATH, "wb") as f:
        f.write(response.content)
    print(f"üìÅ Raw data saved to: {RAW_FILE_PATH}")
    
    # Load the Excel file into a dictionary of DataFrames
    excel_data = pd.read_excel(BytesIO(response.content), sheet_name=None)
    
    # Show available sheets
    print("Sheets found:", list(excel_data.keys()))
    
    # Load main sheet (assuming first sheet is the main one)
    df_main = excel_data[list(excel_data.keys())[0]]
    print("\nFirst 5 rows of the main sheet:")
    print(df_main.head())
    
    # Print headers
    print("\nHeaders in the main sheet:")
    print(df_main.columns.tolist())
else:
    print(f"‚ùå Failed to download. Status Code: {response.status_code}")
```
